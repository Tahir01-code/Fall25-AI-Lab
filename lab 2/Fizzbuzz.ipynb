```python
print(" Lets Begin \n")

previous_number = 0  

for number in range(1, 101):  
    print(f" Number : {number}")

    if previous_number == 0:
        print("No previous number\n")
        previous_number = number
        continue

    total = previous_number + number
    print(f"Total = {previous_number} + {number} = {total}")

    if total % 3 == 0:
        print("Fizz\n")
    elif total % 5 == 0:
        print("Buzz\n")
    elif total % 3 == 0 and total % 5 == 0:
        print("FizzBuzz\n")
    else:
        print(f"{total}\n")
```
